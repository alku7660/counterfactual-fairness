# Synthetic Data Generator:

## Generator function 1:
\begin{equation}
f1(x) = x[1] − 4x[2] + 2x[3] + 3, x[1] ≤ 10 \\
f1(x) = −2x[1] − 3x[2] + x[2] − 2, 10 < x[1] ≤ 20 \\
f1(x) = 3x[1] + x[2] − 2x[3] + 2, x[1] > 20 \\
x[i], \ \epsilon \ [0, 30]
\end{equation}

## Generator function 2:
\begin{equation}
f2(x) = x[1]^3 − 2x[2]^2 + 3x[3] \\ 
x[i], \ \epsilon \ [−100, 100]
\end{equation}

## Generator function 3:
\begin{equation}
f3(x) = x[1] − x[2] sinx[2]^2 \\
x[i], \ \epsilon \ [−10, 10]
\end{equation}

## Generator function 4:
\begin{equation}
f4(x) = ((x[1] == 1) ∧ x[2] ∧ x[3]) ∨ ((x[1] == 2) ∧ x[4] ∧ x[5]) ∨ ((x[1] == 3) ∧ x[6] ∧ x[7]) ∨ ((x[1] == 4) ∧ x[8] ∧ x[9]) \\
x[1], \ \epsilon \ [1, 4], x[2], . . . , x[9] \ \epsilon \ {0,1}
\end{equation}

## Functions

```python
def f_1(x):
    """
    Function used for synthetic dataset 1 and synthetic dataset 2.
    Input x: Array of size 3 or 10 with numerical features between 0 and 30, inclusive.
    Output f1: Value of the function f1(x) according to the synthetic function described in (1)
    """
    if x[0] <= 10:
        f1 = x[0] - 4*x[1] + 2*x[2] + 3
    elif x[0] > 10 and x[0] <= 20:
        f1 = -2*x[0] - 3*x[1] + x[1] - 2
    else:
        f1 = 3*x[0] + x[1] -2*x[2] + 2
    return f1

def f_2(x):
    """
    Function used for synthetic dataset 3
    Input x: Array of size 3 with numerical features between -100 and 100, inclusive.
    Output f2: Value of the function f2(x) according to the synthetic function described in (1)
    """
    f2 = x[0]**3 - 2*x[1]**2 + 3*x[2]
    return f2

def f_3(x):
    """
    Function used for synthetic dataset 4
    Input x: Array of size 2 with numerical features between -10 and 10, inclusive.
    Output f3: Value of the function f3(x) according to the synthetic function described in (1)
    """
    f3 = x[0] - x[1]*math.sin(x[1])**2
    return f3

def f_4(x):
    """
    Function used for synthetic dataset 5 and synthetic dataset 6
    Input x: Array of size 10 or 20 with numerical features between -10 and 10, inclusive.
    Output f3: Value of the function f3(x) according to the synthetic function described in (1)
    """
    f4 = ((x[0] == 1) and x[1] and x[2]) or ((x[0] == 2) and x[3] and x[4]) or ((x[0] == 3) and x[5] and x[6]) or ((x[0] == 4) and x[7] and x[8])
    return f4

def binarization_float(f):
    """
    Function that redefines the output function of the synthetic datasets into the [0,1] values.
    Input f: Value of the function (type float).
    Output f: binarized value of the function.
    """
    if f > 0:
        f = 1
    else:
        f = 0
    return f

def binarization_bool(f):
    """
    Function that redefines the output function of the synthetic datasets into the [0,1] values.
    Input f: Value of the function (type bool).
    Output f: binarized value of the function.
    """
    if f == True:
        f = 1
    else:
        f = 0
    return f

def synthetic1(N):
    """
    Function used to generate the random samples under conditions required for synthetic dataset 1
    Input N: Number of samples to be generated.
    Output synth1: synthetic dataset 1 as list
    """
    synth1 = []
    for i in range(N):
        x = np.random.uniform(0,30,size=(3)).tolist()
        f_1x = f_1(x)
        f_1x = binarization_float(f_1x)
        x.extend([f_1x])
        synth1.append(x)
    return synth1

def synthetic2(N):
    """
    Function used to generate the random samples under conditions required for synthetic dataset 2
    Input N: Number of samples to be generated.
    Output synth2: synthetic dataset 2 as list
    """
    synth2 = []
    for i in range(N):
        x = np.random.uniform(0,30,size=(10)).tolist()
        f_1x = f_1(x)
        f_1x = binarization_float(f_1x)
        x.extend([f_1x])
        synth2.append(x)
    return synth2

def synthetic3(N):
    """
    Function used to generate the random samples under conditions required for synthetic dataset 3
    Input N: Number of samples to be generated.
    Output synth3: synthetic dataset 3 as list
    """
    synth3 = []
    for i in range(N):
        x = np.random.uniform(-100,100,size=(3)).tolist()
        f_2x = f_2(x)
        f_2x = binarization_float(f_2x)
        x.extend([f_2x])
        synth3.append(x)
    return synth3

def synthetic4(N):
    """
    Function used to generate the random samples under conditions required for synthetic dataset 4
    Input N: Number of samples to be generated.
    Output synth4: synthetic dataset 4 as list
    """
    synth4 = []
    for i in range(N):
        x = np.random.uniform(-10,10,size=(2)).tolist()
        f_3x = f_3(x)
        f_3x = binarization_float(f_3x)
        x.extend([f_3x])
        synth4.append(x)
    return synth4

def synthetic5(N):
    """
    Function used to generate the random samples under conditions required for synthetic dataset 5
    Input N: Number of samples to be generated.
    Output synth5: synthetic dataset 5 as list
    """
    synth5 = []
    for i in range(N):
        x1 = np.random.randint(1,5,size=(1)).tolist()
        x2 = np.random.randint(0,2,size=(8)).tolist()
        x1.extend(x2)
        f_4x = f_4(x1)
        f_4x = binarization_bool(f_4x)
        x1.extend([f_4x])
        synth5.append(x1)
    return synth5

def synthetic6(N):
    """
    Function used to generate the random samples under conditions required for synthetic dataset 6
    Input N: Number of samples to be generated.
    Output synth6: synthetic dataset 6 as list
    """
    synth6 = []
    for i in range(N):
        x1 = np.random.randint(1,5,size=(1)).tolist()
        x2 = np.random.randint(0,2,size=(19)).tolist()
        x1.extend(x2)
        f_4x = f_4(x1)
        f_4x = binarization_bool(f_4x)
        x1.extend([f_4x])
        synth6.append(x1)
    return synth6
```

f1(x,y) = N_{x,y}(2,0.5) \\

## Generator function 8:
\begin{equation}
f1(x,y) = N_{x,y}(2,0.5) \\ 
f2(x,y) = N_{x,y}(6,0.5) ∨ (N_{x}(2,0.5),N_{y}(6,0.5)) ∨ (N_{x}(6,0.5),N_{y}(2,0.5))

\begin{equation}
f1(x,y) = [x = 0, N_{y}(2,0.5)] \\ 
f2(x,y) = [x = 1, N_{y}(6,0.5) ∨ N_{y}(2,0.5)] ∨ [x = 0,N_{y}(6,0.5)] 
\end{equation}

## Generator function 12:
\begin{equation}
f1(x,y) = [B_{x}(0,1), N_{y}(4,2)] \\ 
f2(x,y) = [B_{x}(0,1), N_{y}(4,2)] 
\end{equation}

```python
def synthetic7(N):
    """
    Function used to generate the random samples according to generator function 7
    Input N: Number of samples to be generated
    Output synth7: synthetic dataset 7 as list  
    """
    synth7 = []
    for i in range(int(N/2)):
        x0 = np.random.normal(loc=2,scale=0.5,size=2).tolist()
        x0.extend([0])
        x1 = np.random.normal(loc=6,scale=0.5,size=2).tolist()
        x1.extend([1])
        synth7.append(x0)
        synth7.append(x1)
    return synth7

def synthetic8(N):
    """
    Function used to generate the random samples according to generator function 8 
    Input N: Number of samples to be generated
    Output synth8: synthetic dataset 8 as list  
    """
    synth8 = []
    for i in range(int(N/2)):
        x0 = np.random.normal(loc=2,scale=0.5,size=2).tolist()
        x0.extend([0])
        x11 = np.random.normal(loc=6,scale=0.5,size=2).tolist()
        x12 = np.random.normal(loc=2,scale=0.5,size=1).tolist()
        x12_y = np.random.normal(loc=6,scale=0.5,size=1).tolist()
        x12.extend(x12_y)
        x13 = np.random.normal(loc=6,scale=0.5,size=1).tolist()
        x13_y = np.random.normal(loc=2,scale=0.5,size=1).tolist()
        x13.extend(x13_y)
        x1_list = [x11,x12,x13]
        random_sel = np.random.randint(0,3)
        x1 = x1_list[random_sel]
        x1.extend([1])
        synth8.append(x0)
        synth8.append(x1)
    return synth8

def synthetic9(N):
    """
    Function used to generate the random samples according to generator function 9 
    Input N: Number of samples to be generated
    Output synth9: synthetic dataset 9 as list  
    """
    synth9 = []
    for i in range(int(N/2)):
        x0 = np.random.normal(loc=4,scale=2,size=2).tolist()
        x0.extend([0])
        x1 = np.random.normal(loc=4,scale=2,size=2).tolist()
        x1.extend([1])
        synth9.append(x0)
        synth9.append(x1)
    return synth9

def synthetic10(N):
    """
    Function used to generate the random samples according to generator function 10 
    Input N: Number of samples to be generated
    Output synth10: synthetic dataset 10 as list  
    """
    synth10 = []
    for i in range(int(N/2)):
        x0 = [0]
        x01 = np.random.normal(loc=3,scale=0.5,size=1).tolist()
        x0.extend(x01)
        x0.extend([0])
        x1 = [1]
        x11 = np.random.normal(loc=5,scale=0.5,size=1).tolist()
        x1.extend(x11)
        x1.extend([1])
        synth10.append(x0)
        synth10.append(x1)
    return synth10

def synthetic11(N):
    """
    Function used to generate the random samples according to generator function 11 
    Input N: Number of samples to be generated
    Output synth11: synthetic dataset 11 as list  
    """
    synth11 = []
    for i in range(int(N/2)):
        x0 = [0]
        x01 = np.random.normal(loc=2,scale=0.5,size=1).tolist()
        x0.extend(x01)
        x0.extend([0])
        x11 = [1]
        x12 = [1]
        x11_y = np.random.normal(loc=2,scale=0.5,size=1).tolist()
        x12_y = np.random.normal(loc=6,scale=0.5,size=1).tolist()
        x11.extend(x11_y)
        x12.extend(x12_y)
        x13 = [0]
        x13_y = np.random.normal(loc=6,scale=0.5,size=1).tolist()
        x13.extend(x13_y)
        x1_list = [x11,x12,x13]
        random_sel = np.random.randint(0,3)
        x1 = x1_list[random_sel]
        x1.extend([1])
        synth11.append(x0)
        synth11.append(x1)
    return synth11

def synthetic12(N):
    """
    Function used to generate the random samples according to generator function 12 
    Input N: Number of samples to be generated
    Output synth12: synthetic dataset 12 as list  
    """
    synth12 = []
    for i in range(int(N/2)):
        x0 = [np.random.randint(0,2)]
        x01 = np.random.normal(loc=4,scale=2,size=1).tolist()
        x0.extend(x01)
        x0.extend([0])
        x1 = [np.random.randint(0,2)]
        x11 = np.random.normal(loc=4,scale=2,size=1).tolist()
        x1.extend(x11)
        x1.extend([1])
        synth12.append(x0)
        synth12.append(x1)
    return synth12
```
